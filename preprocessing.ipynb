# QuanQual finding

```python
quan=[]
qual=[]
for i in dataset.columns:
    #print(i)
    if dataset[i].dtype=='O':
        qual.append(i)
    else:
        quan.append(i)
```

```python
qual
```

# Finding Outliers

```python
dataset.describe()
```

```python
table=pd.DataFrame(index=['Q:25%','Q:50%','Q:75%','Q:100%','IQR','1.5 rule',
                'lesser outliers','grater outliers','min','max'],columns=quan)
```

```python
for i in quan:
    
    table[i]['Q:25%']=dataset.describe()[i]['25%']
    table[i]['Q:50%']=dataset.describe()[i]['50%']
    table[i]['Q:75%']=dataset.describe()[i]['75%']
    table[i]['Q:100%']=dataset.describe()[i]['max']
    table[i]['IQR']=table[i]['Q:75%']-table[i]['Q:25%']
    table[i]['1.5 rule']=1.5*table[i]['IQR']
    table[i]['lesser outliers']=table[i]['Q:25%']-table[i]['1.5 rule']
    table[i]['grater outliers']=table[i]['Q:75%']+table[i]['1.5 rule']
    table[i]['min']=dataset[i].min()
    table[i]['max']=dataset[i].max()
    
    
    
```

```python
table
```

# Removing outliers

```python
lesser=[]
grater=[]
for i in quan:
    if table[i]['lesser outliers']>table[i]['min']:
        lesser.append(i)
    if table[i]['grater outliers']<table[i]['max']:
        grater.append(i)  
```

```python
lesser
```

```python
grater
```

```python
for i in lesser:
    dataset[i][dataset[i]<table[i]['lesser outliers']]=table[i]['lesser outliers']
```

```python
table=dataset
```
